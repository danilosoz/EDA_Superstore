# Superstore Sales Dataset

Este projeto utiliza o **Superstore Sales Dataset**, que contém dados de vendas de uma empresa fictícia. O conjunto de dados inclui informações detalhadas sobre vendas, lucro, categorias de produtos, subcategorias, regiões, segmentos de clientes e muito mais. 

## Descrição do Dataset

- **Vendas**: Valor total das vendas realizadas.
- **Lucro**: Lucro obtido em cada transação.
- **Categorias de Produtos**: Divisão dos produtos em categorias principais (ex: Móveis, Tecnologia, Papelaria).
- **Subcategorias**: Divisão mais específica dentro de cada categoria.
- **Regiões**: Localização geográfica das vendas (ex: Norte, Sul, Leste, Oeste).
- **Segmentos de Clientes**: Classificação dos clientes (ex: Consumidor, Corporativo, Home Office).
- **Datas**: Período das transações.

## Objetivo do Projeto

O objetivo deste projeto é analisar os dados de vendas para identificar tendências, padrões e insights que possam ajudar a empresa a melhorar suas estratégias de vendas, aumentar a lucratividade e entender melhor o comportamento dos clientes.

```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv("Sample - Superstore.csv", encoding="ISO-8859-1")

# Traduzir as colunas relevantes
traducao_categorias = {
    "Furniture": "Mobília",
    "Office Supplies": "Material de Escritório",
    "Technology": "Tecnologia",
    "West": "Oeste",
    "East": "Leste",
    "South": "Sul",
    "Central": "Centro"
}

traducao_subcategorias = {
    "Bookcases": "Estantes",
    "Chairs": "Cadeiras",
    "Labels": "Etiquetas",
    "Tables": "Mesas",
    "Storage": "Armazenamento",
    "Furnishings": "Decoração",
    "Art": "Arte",
    "Phones": "Telefones",
    "Binders": "Encadernadores",
    "Appliances": "Eletrodomésticos",
    "Paper": "Papel",
    "Accessories": "Acessórios",
    "Envelopes": "Envelopes",
    "Fasteners": "Fixadores",
    "Supplies": "Suprimentos",
    "Machines": "Máquinas",
    "Copiers": "Copiadoras"
}

# Aplicar tradução ao DataFrame
df["Category"] = df["Category"].map(traducao_categorias).fillna(df["Category"])
df["Region"] = df["Region"].map(traducao_categorias).fillna(df["Region"])
df["Sub-Category"] = df["Sub-Category"].map(traducao_subcategorias).fillna(df["Sub-Category"])

# Visualizar as primeiras linhas
df.head()
```

```python
# Verificar as colunas
print(df.columns)

# Verificar informações gerais
df.info()

# Ajustar a formatação para cada coluna
pd.set_option('display.float_format', '{:.2f}'.format)  # Padrão para a maioria das colunas
df_desc = df.describe()

# Ajustes específicos
df_desc["Quantity"] = df_desc["Quantity"].apply(lambda x: round(x, 0))  # Quantidade sem decimais
df_desc["Row ID"] = df_desc["Row ID"].apply(lambda x: round(x, 0))      # Row ID sem decimais
df_desc["Postal Code"] = df_desc["Postal Code"].apply(lambda x: round(x, 0))  # Postal Code sem decimais

# Estatísticas descritivas
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar vendas por categoria
vendas_por_categoria = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(
    x=vendas_por_categoria.index,       # Categorias no eixo x
    y=vendas_por_categoria.values,      # Vendas no eixo Y
    hue=vendas_por_categoria.index,     # Usar 'x' como 'hue' para aplicar a paleta
    palette="viridis",                  # Paleta de cores
    dodge=False,                        # Evitar que as barras sejam separadas
    legend=False                        # Não mostrar a legenda
)
plt.title("Vendas por Categoria", fontsize=16)
plt.xlabel("Categoria", fontsize=12)
plt.ylabel("Vendas Totais (em USD)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#Agrupar lucro por subcategoria
lucro_por_subcategoria = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)

# Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(
    x=lucro_por_subcategoria.index,       # Subcategorias no eixo x
    y=lucro_por_subcategoria.values,      # Lucro no eixo Y
    hue=lucro_por_subcategoria.index,     # Usar 'x' como 'hue' para aplicar a paleta
    palette="magma",                      # Paleta de cores
    dodge=False,                          # Evitar que as barras sejam separadas
    legend=False                          # Não mostrar a legenda
)
plt.title("Lucro por Subcategoria", fontsize=16)
plt.xlabel("Subcategoria", fontsize=12)
plt.ylabel("Lucro Total (em USD)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Agrupar vendas e lucro por região
vendas_lucro_por_regiao = df.groupby("Region").agg({"Sales": "sum", "Profit": "sum"})

# Gráfico de barras (vendas)
plt.figure(figsize=(10, 6))
sns.barplot(
    x=vendas_lucro_por_regiao.index,       # Regiões no eixo x
    y=vendas_lucro_por_regiao["Sales"],    # Vendas no eixo Y
    hue=vendas_lucro_por_regiao.index,     # Usar 'x' como 'hue' para aplicar a paleta
    palette="coolwarm",                    # Paleta de cores
    dodge=False,                           # Evitar que as barras sejam separadas
    legend=False                           # Não mostrar a legenda
)
plt.title("Vendas por Região", fontsize=16)
plt.xlabel("Região", fontsize=12)
plt.ylabel("Vendas Totais (em USD)", fontsize=12)
plt.show()

# Gráfico de barras (lucro)
plt.figure(figsize=(10, 6))
sns.barplot(
    x=vendas_lucro_por_regiao.index,       # Regiões no eixo x
    y=vendas_lucro_por_regiao["Profit"],   # Lucro no eixo Y
    hue=vendas_lucro_por_regiao.index,     # Usar 'x' como 'hue' para aplicar a paleta
    palette="coolwarm",                    # Paleta de cores
    dodge=False,                           # Evitar que as barras sejam separadas
    legend=False                           # Não mostrar a legenda
)
plt.title("Lucro por Região", fontsize=16)
plt.xlabel("Região", fontsize=12)
plt.ylabel("Lucro Total (em USD)", fontsize=12)
plt.show()
```

## Insights e Conclusões

- A categoria **Tecnologia** é a que gera mais vendas, enquanto **Material de Escritório** tem o menor volume de vendas.
- A subcategoria **Copiadoras** é a mais lucrativa, enquanto **Mesas** tem o menor lucro.
- A região **Oeste** é a que mais vende e também a mais lucrativa.
- Descontos altos podem estar impactando negativamente o lucro em algumas subcategorias.
